```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

datagen = ImageDataGenerator(
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
for i in range(90):
  img = load_img("/content/drive/MyDrive/Minor_Project/Dataset/A/subA/frame"+str(i)+".jpg")
  x = img_to_array(img)
  x = x.reshape((1,) + x.shape)

i=0
for batch in datagen.flow(x, batch_size=1, save_to_dir="/content/drive/MyDrive/Minor_Project/Dataset/A/subA", save_prefix='copy', save_format='jpg'):
  i += 1
  if(i>=110):
    break
```
