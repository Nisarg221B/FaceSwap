```python
import cv2
from facenet_pytorch import MTCNN
from PIL import Image
import torch
import matplotlib.pyplot as plt
import torchvision.transforms as T
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
mtcnn = MTCNN(keep_all=True, device=device, margin=50, select_largest=True, image_size=256)
```

```python
video = cv2.VideoCapture("/content/drive/MyDrive/Minor_Project/Dataset/videos/elon.mp4")

currentframe = 1
while(True):
  success, frame = video.read()
  if success:
    frame = Image.fromarray(frame)
    boxes, _ = mtcnn.detect(frame)
    for box in boxes:
        box_list = box.tolist()
        # bounding box coordinated
        x1 = int(box_list[0])
        y1 = int(box_list[1])
        x2 = int(box_list[2])
        y2 = int(box_list[3])
        #  find the middle of the image to get a perfect square, mtcnn gives a rectangle image of the face so making
        #  the image a square makes it easier to train
        y1 += 5
        y2 -= 5
        diff = abs(y1 - y2)
        mid_x = (x2 + x1) // 2
        # mid_y = (y2 + y1) // 2
        x1 = mid_x - (diff // 2)
        x2 = mid_x + (diff // 2)
        f_final = frame.crop((x1, y1, x2, y2))  # sends back only the square around the face, possible no face detected
        f_final.save("/content/drive/MyDrive/Minor_Project/Dataset_2/A/subA"+"/{}.png".format(str(currentframe).zfill(4)))
    currentframe += 1
  else:
    break

video.release()
cv2.destroyAllWindows()
```
